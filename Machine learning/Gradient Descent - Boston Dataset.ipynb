```python
import numpy as np
import pandas as pd
boston_data=pd.read_csv("Datasets/train.csv")
boston_data
```

```python
x_train_boston_data = boston_data.iloc[:,0:13]
x_train_boston_data
```

```python
y_train_boston_data= boston_data.iloc[:,1]
y_train_boston_data
```

```python
from sklearn.ensemble import GradientBoostingRegressor
model = GradientBoostingRegressor()
model.fit(x_train_boston_data,y_train_boston_data)
boston_data1 = np.loadtxt("Datasets/test.csv" , delimiter=',')
print(boston_data1)
print(boston_data1.shape)
```

```python
y_predict = model.predict(boston_data1)
y_predict
```

```python
np.savetxt("Datasets/boston_predict.csv", y_predict)
```
