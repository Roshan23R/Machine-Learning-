```python
import numpy as np
import matplotlib.pyplot as plt
# color, marker, linewidth, labels for axis, datasets and the title, axis, grid, text
#x = np.array([1,2,3,4])
x = np.arange(0, 5, 0.1)
y  = x **3
y2 = x**2
#plt.plot(x, y, color = "black", marker="p")
plt.plot(x, y, color = "black", linewidth = 5, label = "x^3")

plt.plot(x, y2, color = "red", linewidth = 5, label = "x^2")

# plt.axis([0, 6, 0, 150])

plt.ylabel("y")
plt.xlabel("x")

plt.grid()

plt.text(2, 80, "Text", fontsize=12)

plt.legend()
plt.title("Matplotlib demo")
plt.show()
```

# Bubble Graphs

```python
import matplotlib.pyplot as plt
```

```python
year=[2012,2013,2014,2015,2016,2017]
salary=[12,13,11,15,18,19]
population=[150,130,100,200,250,300]
```

```python
plt.scatter(year, salary, s=population , c='r' , alpha=0.5)
plt.title("Bubble Chart")
plt.show()
```

```python
import numpy as np
c=np.arange(len(year))
```

```python
plt.scatter(year, salary, s=population , c=c , edgecolor='black', alpha=0.5)
plt.text(year[0]+0.1,salary[0]+0.1,population[0])
plt.title("Bubble Chart")
plt.show()
```

# PIE CHART 

```python
sizes=[3,4,6,2]
labels=["A","B","C","D"]
colors=["blue","gold","maroon","green"]
explode=[0.1,0,0,0]
plt.pie(sizes, colors=colors, labels=labels, explode=explode, autopct="%.2f" , counterclock=False)
plt.axis("equal")
plt.show()
```
